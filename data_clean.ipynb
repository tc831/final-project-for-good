```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
data_file = "Resources/matches_2.csv"
data_file_df = pd.read_csv(data_file)
data_file_df
```

```python
matches_df = data_file_df.drop(columns=['Unnamed: 0', 'notes'], axis=1)
matches_df = matches_df.rename(columns={'match report': 'match_report'})
```

```python
matches_df
```

```python
matches_final = matches_df.drop_duplicates(subset=['date', 'time', 'comp', 'round', 'day', 'venue', 'result',
       'gf', 'ga', 'opponent', 'xg', 'xga', 'poss', 'attendance', 'captain',
       'formation', 'referee', 'match_report', 'sh', 'sot', 'dist',
       'fk', 'pk', 'pkatt', 'season', 'team'])
```

```python
# Setup Connection to Postgres
from config import password

protocol = 'postgresql'
username = 'postgres'
host = 'localhost'
port = 5432
database_name = 'final_project'


rds_connection_string = f'{protocol}://{username}:{password}@{host}:{port}/{database_name}'
engine = create_engine(rds_connection_string)
engine.table_names()
```

```python
# Importing Data to Postgres
matches_final.to_sql('matches_final', con=engine, if_exists='replace', index=False)
```

```python
# Testing data import
pd.read_sql_query('select * from matches_final', con=engine)
```
