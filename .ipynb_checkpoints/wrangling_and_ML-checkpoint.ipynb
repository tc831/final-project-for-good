# 2. Check column values to numerical

```python
def day_converter(x):
    if x == "Mon":
        return 0
    if x == "Tue":
        return 1
    if x == "Wed":
        return 2
    if x == "Thu":
        return 3
    if x == "Fri":
        return 4
    if x == "Sat":
        return 5
    if x == "Sun":
        return 6
 
```

```python
file["day"] = file["day"].map(lambda x: day_converter (x))   
file["day"].astype('int64')
file.head()
```

```python
def location_converter(x):
    if x == "Home":
        return 0
    if x == "Away":
        return 1
    
file["venue"] = file["venue"].map(lambda x: location_converter (x))   
file["venue"].astype('int64')
file.head()
```

```python
def result_converter(x):
    if x == "L":
        return 0
    if x == "D":
        return 1
    if x == "W":
        return 2
    
file["result"] = file["result"].map(lambda x: result_converter (x))   
file["venue"].astype('int64')
file.head()
```

## Pick the columns which 

```python
file_corr = file[['day', 'venue','result', 'gf', 'ga', 'opponent','xg', 'xga', 'poss',
                 'attendance', 'sh', 'sot', 'dist', 'fk', 'pk', 'pkatt','season', 'team']]
```

```python
names = file_corr.columns
data = file
correlations = data.corr()
fig = pyplot.figure(figsize=(19,16))
ax = fig.add_subplot(111)
cax = ax.matshow(correlations, vmin=-1, vmax=1)
fig.colorbar(cax)

ticks = numpy.arange(0,18,1)
ax.set_xticks(ticks)
ax.set_yticks(ticks)
ax.set_xticklabels(names)
ax.set_yticklabels(names)
pyplot.show()
# >0.5 weak corrleation
```

```python
X = file_corr.drop(["result", "attendance"], axis =1)
y = file_corr["result"].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
X_train.head()
```

```python
import seaborn as sns
plt.figure(figsize=(15,15))
cor = X_train.corr()
sns.heatmap(cor, annot=True, cmap=plt.cm.CMRmap_r)
```

## Define a function to select highly correlated features but remove the first feature that is correlated with other feature

```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = file_corr.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold:
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
        return col_corr
```

```python
corr_features = correlation(X_train, 0.7)
len(set(corr_features))
```
