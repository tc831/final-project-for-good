```python
from splinter import Browser
from bs4 import BeautifulSoup as bs
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)

base_url = 'https://fbref.com/'

scores_url = 'https://fbref.com/en/comps/9/schedule/Premier-League-Scores-and-Fixtures'

browser.visit(scores_url)

time.sleep(1)

# Scrape page into Soup
html = browser.html
soup = bs(html, "html.parser")
```

```python
# next_season_comps = soup.find("div", class_='comps')
next_season = soup.find("a", class_="next")
next_season_link = base_url + next_season["href"]
browser.quit()
```

```python
table = pd.read_html(next_season_link)
```

```python
upcoming_matches = table[0]
upcoming_matches = upcoming_matches[['Wk', 'Day', 'Date', 'Time', 'Home', 'Away', 'Venue']]
```

```python
upcoming_matches
```

```python
upcoming_matches_df = upcoming_matches.iloc[1:]
```

```python
upcoming_matches_df = upcoming_matches_df.reset_index(drop=True)
```

```python
import numpy as np
upcoming_matches_table = []
for i in np.arange(len(upcoming_matches_df)):
    up_wk = upcoming_matches_df['Wk']
    up_day = upcoming_matches_df['Day']
    up_date = upcoming_matches_df['Date']
    up_time = upcoming_matches_df['Time']
    up_home = upcoming_matches_df['Home']
    up_away = upcoming_matches_df['Away']
    up_venue = upcoming_matches_df['Venue']
    upcoming_matches_table.append({'Wk': up_wk, 'Day': up_day, 'Date': up_date, 'Time': up_time, 'Home': up_home, 'Away': up_away, 'Venue': up_venue})
upcoming_matches_table
```
